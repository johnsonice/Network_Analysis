```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import operator
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
```

```python
G = nx.karate_club_graph()
```

### Show degree distribution

```python
degrees = G.degree()
degree_values = [i[1] for i in degrees]
```

```python
plt.hist(x=degree_values,bins='auto',rwidth=0.85)#,color='#607c8e')
plt.title('Drgree distribution')
plt.ylabel('Frequency of nodes')
plt.xlabel('Degrees')
plt.show()
```

In real world, degree distribution is usually in log log scale. Most of the nodes have few degree. But a few nodes have very high degree. 

It is usually called power law

### Preferential Attachment Model
- a network generate process which follows power law

barabasi_albert_graph(n, m) returns a network with n nodes. Each new node attaches to m
existing nodes according to the Preferential Attachment model. 

```python
G = nx.barabasi_albert_graph(n=10000,m=1)
```

```python
degrees = G.degree()
degree_values = [i[1] for i in degrees]
plt.hist(x=degree_values,bins=100)#,color='#607c8e')
plt.title('Drgree distribution')
plt.ylabel('Frequency of nodes')
plt.xlabel('Degrees')
plt.show()
```

```python
plt.hist(x=degree_values,bins = 100, histtype='step')#,color='#607c8e')
plt.title('Drgree distribution in log log scale')
plt.ylabel('Frequency of nodes')
plt.xlabel('Degrees')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
from collections import Counter
from operator import itemgetter
```

```python
his = Counter(degree_values)
his = sorted(his.items(),key=itemgetter(0)) 
x = [i[0] for i in his]
y = [i[1]/len(G.nodes) for i in his]
```

```python
plt.plot(x,y, 'o')
plt.title('Drgree distribution in log log scale')
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
